```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer
from umap import UMAP
import matplotlib.pyplot as plt

# Prepare the data
side_effects_reviews = final_result['sideEffectsReview'].tolist()

# Define embedding models
embedding_models = {
    "all-MiniLM-L6-v2": SentenceTransformer("all-MiniLM-L6-v2"),
    "all-mpnet-base-v2": SentenceTransformer("all-mpnet-base-v2"),
  #  "paraphrase-multilingual-MiniLM-L12-v2": SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
}

# Function to create and fit BERTopic model
```

```python
def create_bertopic_model(embedding_model, docs):
    # Preprocess the docs
    docs = [str(doc) for doc in docs]
    docs = [doc if doc != 'nan' else '' for doc in docs]
    
    vectorizer = CountVectorizer(stop_words="english")
    topic_model = BERTopic(
        embedding_model=embedding_model,
        vectorizer_model=vectorizer,
        nr_topics="auto"
    )
    topics, probs = topic_model.fit_transform(docs)
    return topic_model, topics, probs
```

```python
import random
import numpy as np

# Set seeds for reproducibility
random.seed(42)
np.random.seed(42)

# Fit models and store bertopic_results
bertopic_results = {}
for model_name, embedding_model in embedding_models.items():
    topic_model, topics, probs = create_bertopic_model(embedding_model, side_effects_reviews)
    bertopic_results[model_name] = {
        "model": topic_model,
        "topics": topics,
        "probs": probs
    }

# Compare models
for model_name, result in bertopic_results.items():
    print(f"\nModel: {model_name}")
    print(result["model"].get_topic_info())
```

```python
# Visualize document clustering for each model
for model_name, result in bertopic_results.items():
    embeddings = result["model"].embedding_model.embed(side_effects_reviews)
    umap_model = UMAP(n_neighbors=15, n_components=2, min_dist=0.1, metric='cosine')
    embeddings_2d = umap_model.fit_transform(embeddings)

    plt.figure(figsize=(10, 8))
    plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=result["topics"], cmap='hsv', alpha=0.1)
    plt.colorbar()
    plt.title(f'Document Clusters - {model_name}')
    plt.savefig(f"./Clusters_images/{model_name}_clusters.png")
    plt.close()
```

```python
# Calculate topic diversity
def calculate_topic_diversity(model, top_n=10):
    topics = model.get_topics()
    unique_words = set()
    for topic in topics:
        unique_words.update([word for word, _ in topic[:top_n]])
    return len(unique_words) / (len(topics) * top_n)

for model_name, result in bertopic_results.items():
    diversity = calculate_topic_diversity(result["model"])
    print(f"{model_name} Topic Diversity: {diversity}")
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

def extract_patient_concerns(patient_paragraph):
    vectorizer = CountVectorizer(stop_words='english', max_df=1.0, min_df=1)
    X = vectorizer.fit_transform([patient_paragraph])
    lda = LatentDirichletAllocation(n_components=1, random_state=0)
    lda.fit(X)
    words = vectorizer.get_feature_names_out()
    components = lda.components_
    keywords = [words[i] for i in components.argsort()[0][-5:][::-1]]
    return keywords

def filter_drugs_based_on_patient_side_effects(bertopic_results, patient_side_effects):
    filtered_drugs = []
    
    for drug, models in bertopic_results.items():
        model_result = models['all-MiniLM-L6-v2']
        topic_info = model_result['model'].get_topic_info()
        
        match_found = False
        for _, row in topic_info.iterrows():
            topic_words = row['Name'].split('_')[1:]  # Exclude the topic number
            topic_words = [word.lower() for word in topic_words if isinstance(word, str)]
            if any(effect.lower() in topic_words for effect in patient_side_effects):
                match_found = True
                break
        
        if not match_found:
            filtered_drugs.append(drug)
    
    return filtered_drugs

# Example usage
patient_paragraph = "I do not want to experience dry skin, headache, or nausea while using these medications. I'm also concerned about potential hair loss and mood changes."
patient_concerns = extract_patient_concerns(patient_paragraph)

print("Extracted patient concerns:", patient_concerns)

# Use these concerns in the drug filtering function
recommended_drugs = filter_drugs_based_on_patient_side_effects(bertopic_results, patient_concerns)

print("\nRecommended drugs (side effects not matching patient's concerns):")
for drug in recommended_drugs:
    print(f"- {drug}")

print("\nFiltered out drugs (potential side effects matching patient's concerns):")
filtered_out = set(bertopic_results.keys()) - set(recommended_drugs)
for drug in filtered_out:
    print(f"- {drug}")
```
